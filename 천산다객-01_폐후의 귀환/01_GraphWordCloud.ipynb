```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from matplotlib import font_manager, rc
import numpy as np
import pandas as pd
from konlpy.tag import Okt
from collections import Counter
from PIL import Image
import random
from gensim import corpora
import gensim
import pyLDAvis
import pyLDAvis.gensim_models
import nltk
```

```python
#%% [markdown]
#   
webnovel_title = '- '         #    
Graph_number = 50                 # 50   100  .

#  
del_list = ['', '', '', '', '', '', '', '', '', 
            '', '', '', '', '', '', '', '',
            '', '', '', '', '', '', '', '', '',
            '', '', '', '', '', '', '', '', '',
            '', '', '', '', '']

#   (      )
word_mapped = {"": "","": "","": "",
               "": "", "": "","": "", "": ""}

#   
font_path = "c:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

#%% [markdown]
#    
df = pd.read_excel(f'{webnovel_title}-Total_collected_data.xlsx', na_values=['NaN'])  #  
df = df.dropna()  # NaN    
df.reset_index(drop=True, inplace=True)  #  

#     
articles = df['Review'].tolist()
print("Total articles:", len(articles))

#    
articles = ' '.join(articles)
articles = articles[:1000000000]  #     (  )

#     
tokenizer = Okt()
nouns = tokenizer.nouns(articles)  #  

#     
word_cleaned = []
for word in nouns:
    if word in word_mapped:
        word = word_mapped[word]  # '' '' 
    if len(word) > 1 and word not in del_list:
        word_cleaned.append(word)

#  
result = Counter(word_cleaned)
word_dic = dict(result)
sorted_word_dic = sorted(word_dic.items(), key=lambda x: x[1], reverse=True)

#   
word_frequency = nltk.FreqDist(word_cleaned)
df_freq = pd.DataFrame(list(word_frequency.values()), word_frequency.keys())

#  NN    
result = df_freq.sort_values([0], ascending=False)
result = result[:Graph_number]  #  Graph_number   
result = result.iloc[::-1]  #    

#    
if Graph_number == 50:
    result.plot(kind='barh', legend=False, figsize=(10, 14), color='navy')  
elif Graph_number == 100:
    result.plot(kind='barh', legend=False, figsize=(10, 20), color='navy')

plt.title(webnovel_title, fontsize=24)  #     

# x  
max_value = result[0].max()
tick_spacing = (max_value // 4 // 100) * 100  # 5    
plt.gca().set_xticks(np.arange(0, max_value + tick_spacing, tick_spacing)) 

plt.grid(True, which='major', axis='x', linestyle='--', linewidth=1.0)
plt.tick_params(axis='both', which='major', labelsize=12)  # x,y   
plt.tight_layout()

#    
plt.savefig(f'{webnovel_title}-Graph-{Graph_number}.png', dpi=300, bbox_inches='tight')
plt.show()
plt.close()
#%% [markdown]
# WordCloud 
#   
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    colors = ['#008080', '#000080']  # teal navy
    return random.choice(colors)

#   
x, y = np.ogrid[:800, :800]
mask = (x - 400) ** 2 + (y - 400) ** 2 > 400 ** 2
mask = 255 * mask.astype(int)

#   
wc = WordCloud(
    font_path=font_path,
    background_color='white',
    mask=mask,
    width=800,
    height=800,
    max_words=200,
    max_font_size=180,
    color_func=color_func,  #   
    prefer_horizontal=0.9,
    random_state=42)

wc.generate_from_frequencies(word_dic)    #   

#   
plt.figure(figsize=(10, 10))
plt.title(webnovel_title, fontsize=24)  #     
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig(f'{webnovel_title}-.png', dpi=300, bbox_inches='tight')
plt.show()
plt.close()
```
